```python
import numpy
import matplotlib.pyplot as plt
```

```python
a = numpy.load('bpt1.npy')
a.shape
```

```python
plt.figure(figsize = (15,7))
x = a[:,0]
y = a[:,1]
to_nan = numpy.logical_or.reduce([x<-1.6, x>0.5, y>1.4, y<-1.1])
x[to_nan] = numpy.nan
y[to_nan] = numpy.nan
plt.scatter(x,y,s=1)
plt.show()
```

```python
numpy.save('../colors/NII_Ha.npy', x)
numpy.save('../colors/OIII_Hb.npy', y)
```
